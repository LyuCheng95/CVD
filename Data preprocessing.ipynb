```python
def get_healthy(file):
    data = pd.read_csv(file)
    data = data.loc[(data.smoke == 0) & (data.alco == 0) & (data.active == 1)]
    data = data.drop(columns=['smoke','alco','active'])
    print(data.shape)
    filename = file.split(".")
    data.to_csv(filename[0]+"-healthy.csv", index=False)

files = ['t1.csv','t2.csv','t3.csv','t4.csv']

for file in files:
    get_healthy(file)
```

```python
def get_non_smokers(file):
    data = pd.read_csv(file)
    data = data.loc[data.smoke == 0]
    data = data.drop(columns = ["smoke"])
    filename = file.split(".")
    print(data.shape)
    data.to_csv(filename[0]+"-non-smokers.csv", index=False)

for file in files:
    get_non_smokers(file)
```

```python
def split_gender(file):
    data = pd.read_csv(file)
    data1 = data.loc[data.gender == 1]
    data1 = data1.drop(columns = ["gender"])
    data0 = data.loc[data.gender == 0]
    data0 = data0.drop(columns = ["gender"])
    filename = file.split(".")
    print(data1.shape)
    data1.to_csv(filename[0]+"-only-females.csv", index=False)
    data0.to_csv(filename[0]+"-only-males.csv", index=False)
```

```python
for file in files:
    split_gender(file)
```

```python
def non_exercises(file):
    data = pd.read_csv(file)
    data = data.loc[data.active == 0]
    data = data.drop(columns=['active'])
    filename = file.split(".")
    print(data.shape)
    data.to_csv(filename[0]+"-non-exercises.csv", index=False)
    
for file in files:
    non_exercises(file)
```

```python
def get_unhealthy(file):
    data = pd.read_csv(file)
    data = data.loc[(data.smoke == 1) | (data.alco == 1) | (data.active == 0)]
    data = data.drop(columns=['smoke','alco','active'])
    print(data.shape)
    filename = file.split(".")
    data.to_csv(filename[0]+"-unhealthy.csv", index=False)

files = ['t1.csv','t2.csv','t3.csv','t4.csv']

for file in files:
    get_unhealthy(file)
```
