```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

data = pd.read_csv("./new_cardio.csv", sep=",")

data_0 = data[data.cardio==0]
data_1 = data[data.cardio==1]
                
names = data.columns.tolist()

for j in range(len(names)):
    
    if "chol" not in names[j] and "gluc" not in names[j]:

        plt.figure(figsize=(8,8))

        h1 = data_0.iloc[:,j]
        h1mean = np.mean(h1)
        h1std = np.std(h1)
        h1 = sorted(list(h1))

        h2 = data_1.iloc[:,j]
        h2mean = np.mean(h2)
        h2std = np.std(h2)
        h2 = sorted(list(h2))

        print(j)
        print("\t", "mean-diff:", abs(round(h1mean-h2mean,3)), "class0-std:", round(h1std,3), "class1-std:", round(h2std,3))

        if j < 5:
            pdf = stats.norm.pdf(h1, h1mean, h1std)
            plt.plot(h1, pdf, color='blue', label='class 0')
            pdf2 = stats.norm.pdf(h2, h2mean, h2std)
            plt.plot(h2, pdf2, color='red', label='class 1')
            plt.xlabel("Normalised " + names[j])
            plt.ylabel("Qty")

        else:
            plt.hist(h1, bins=[0,1,2] ,color='blue', label='class 0', histtype='step')
            plt.hist(h2, bins=[0,1,2], color='red', label='class 1', histtype='step')
            plt.xlabel(names[j])
            plt.ylabel("Qty")

        plt.legend()
        plt.show()
```

```python
for i in [(0,0,"super high"), (0,1, "abv normal"), (1,0, "normal")]:
    plt.figure(figsize=(8,8))

    h1 = data_0.loc[(data.cholesterol_0 == i[0]) & (data.cholesterol_1 == i[1])].cardio
    h1mean = np.mean(h1)
    h1std = np.std(h1)
    # h1 = sorted(list(h1))

    h2 = data_1.loc[(data.cholesterol_0 == i[0]) & (data.cholesterol_1 == i[1])].cardio
    h2mean = np.mean(h2)
    h2std = np.std(h2)
    # h2 = sorted(list(h2))

    # print(i[0])
    print("\t", "mean-diff:", round(h1mean-h2mean,3), "class0-std:", round(h1std,3), "class1-std:", round(h2std,3))

    plt.hist(h1, bins=[0,1,2] ,color='blue', label='class 0', histtype='step')
    plt.hist(h2, bins=[0,1,2], color='red', label='class 1', histtype='step')
    plt.xlabel("Cholesterol " + i[2] + " (chol_0 = " + str(i[0]) + " and chol_1 = " + str(i[1]) + ")")
    plt.ylabel("Qty")

    plt.legend()
    plt.show()
```
